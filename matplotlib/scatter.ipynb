```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import platform

# 한글 깨짐 방지
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 승차 상위 10
df_top_boarding = df.sort_values(by="승차_승객_수", ascending=False).head(10)
# 하차 상위 10
df_top_alighting = df.sort_values(by="하차_승객_수", ascending=False).head(10)

plt.figure(figsize=(16,6))
colors = plt.cm.tab10.colors

# 왼쪽 subplot: 승차 상위 10
plt.subplot(1, 2, 1)
for i, line in enumerate(df_top_boarding["호선_명칭"]):
    temp = df_top_boarding[df_top_boarding["호선_명칭"] == line]
    plt.scatter(
        temp["승차_승객_수"],
        temp["하차_승객_수"],
        s=temp["승객_수"]/5000,
        color=colors[i % 10],
        alpha=0.7,
        label=line
    )
plt.title("승차 승객 수 상위 10 호선", fontsize=14)
plt.xlabel("승차 승객 수", fontsize=12)
plt.ylabel("하차 승객 수", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# 오른쪽 subplot: 하차 상위 10
plt.subplot(1, 2, 2)
for i, line in enumerate(df_top_alighting["호선_명칭"]):
    temp = df_top_alighting[df_top_alighting["호선_명칭"] == line]
    plt.scatter(
        temp["승차_승객_수"],
        temp["하차_승객_수"],
        s=temp["승객_수"]/5000,
        color=colors[i % 10],
        marker='^',  # 삼각형 마커
        alpha=0.7,
        label=line
    )
plt.title("하차 승객 수 상위 10 호선", fontsize=14)
plt.xlabel("승차 승객 수", fontsize=12)
plt.ylabel("하차 승객 수", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
