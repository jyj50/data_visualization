```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import platform
```

```python
df = pd.read_csv(r'C:\Users\i7 8700K\Desktop\data_visualization\dataset\bar\한국도로교통공단_사고유형별 교통사고 통계_20241231.csv', encoding='cp949')
```

```python
df.info()
```

```python
# 한글 깨짐 방지
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 사고유형대분류별 색상 매핑
color_map = {
    "차대사람": "tomato",
    "차대차": "skyblue",
    "차량단독": "lightgreen",
    "철길건널목": "orange"
}

colors = df["사고유형대분류"].map(color_map)

plt.figure(figsize=(14,6))
plt.bar(df["사고유형"], df["사고건수"], color=colors, edgecolor='black')

plt.title("사고유형별 사고건수 (대분류 색상 구분)", fontsize=16)
plt.xlabel("사고유형", fontsize=12)
plt.ylabel("사고건수", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 범례 추가
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=color, label=cat) for cat, color in color_map.items()]
plt.legend(handles=legend_elements, title="사고유형대분류")

plt.show()

```
